# Префиксное дерево (Trie)

***Определение.*** Префиксное дерево (*Trie*, бор) $-$ структура данных, позволяющая эффективно хранить набор строк.

***Пимер.*** Составим префиксное дерево для следующего набора строк:
- a
- aaa
- aab
- acc
- acb
- ba
- baac
- baab

# Алгоритм

Алгоритм основан на составлении конечного автомата используя префиксное дерево, составленное из набора искомых строк.

***Определение.*** Назовем суффиксной ссылкой дополнительное ребро в префиксном дереве, направленное из вершины в вершину, являющуюся максимальным суффиксом данной вершины (считаем, что каждая вершина префиксного дерева является строкой, составляемой из символов, встречающихся на пути к данной вершине из корня).

Построим суффиксные ссылки в изображенном выше префиксном дереве.

![](./images/trie_2_2.png)

Для построения суффиксных ссылок необходимо сделать один обход в ширину по посроенному префиксному дереву.

Для корня и потомков корня суффиксными ссылками является корень. Для того, чтобы построить суффиксные ссылки для любой другой вершины, необходимо совершить переход к ее родителю, и переходить по суффиксным ссылкам до тех пор, пока не встретится вершина, имеющая переход по той же букве, что и родительская вершина к текущей. Вершина, находящаяся по данному переходу и является суффиксной ссылкой текущей вершины.

Оценим сложность построения суффиксных ссылок.

$d_i = \text{depth}(v_i.\text{suf_link})$ $-$ глубина $i$-ой вершины, это глубина, в которой мы окажемся после перехода по суффиксной ссылке.

$d_i \leq d_{i-1} - k + 1$ $-$ глубина $i$-ой вершины меньше или равна глубине предыдущей вершины, минус все переходы по суффиксным ссылкам, плюс подъем на одну вершину вверх.

$d_n \leq d_0 - \sum \limits_i k_i + n$ $-$ рекурентно раскрываем предыдущую формулу.

$\sum \limits_i k_i \leq n + d_0 - d_n \leq n$

![](./images/trie_3.png)

Таким образом суммарное время построения суффиксных ссылок:

$T[\text{build}] \leq \sum |s_i|$

После построения суффиксных ссылок начнем переходить по вершинам построенного суффиксного дерева по символам строки. Если из текущей вершины нет перехода по текущему символу, то совершаем переход по суффиксным ссылкам до тех пор, пока не попадем в корень или не найдем вершину, имеющую данный переход. Для каждой вершины будем считать, сколько раз она была посещена.

Пусть дана строка *ababaacbaab*

После прохода в вершины будут посущены следующее число раз.

![](./images/trie_5.png)

Теперь просуммируем полученные числа с числами тех вершин, из которых можно совершить переход в текущую по суффиксной ссылке.

![](./images/trie_6.png)

Таким образом в терминальных вершинах мы имеем число раз, которое были найдены строки, соответствующие данным вершинам.
