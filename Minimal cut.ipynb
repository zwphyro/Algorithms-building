# Алгоритм Каргера

Выбираем случайное ребро и стягиваем его следующим образом: вершины, инцидентные данному ребру удаляются, а все ребра, что входили в обе данные вершины становятся инцидентны новой вершине.

Выполнив данную процедуру $(n-2)$ раза получим 2 вершины, соединенные нестяннутыми ребрами, данное множество ребер образует разрез.

Пусть $E$ $-$ множество ребер графа, $V$ $-$ множество вершин в графе, $C$ $-$ множество ребер, входящих в минимальный разрез.

Вероятность того, что на первом шаге было стянуто ребро, входящее в минимальный разрез $-$ $|C| \over |E|$

Пусть $|V| = n, |C| = k$.

Так как из каждой вершины выходит число ребер, больше или равное минимаьному разрезу графа, следовательно число ребер графа можно оценить следующим образом, используя теорему Эйлера:
$$ |E| \geq {kn \over 2} $$

Тогда вероятность что ребро, входящее в минимальный разрез было стянуто на первом шаге:
$$ {|C| \over |E|} \geq {k \over {kn \over 2}} = {2 \over n} $$

Таким образом вероятность того, что стянутое ребро не входит в минимальный разрез $ \leq 1 - {2 \over n} = {n - 2 \over n}$

Тогда алгоритм сработает верно с вероятностью 
$$ \prod\limits_{i=1}^{n-2}P(x_i) \geq {n-2 \over n}{n-3 \over n-1}\cdot... = {(n-2)! \over n!} = {2 \over n(n-1)} \text{, где}$$
$P(x_i)$ $-$ вероятность, что алгоритм отработал корректно на $i$-ом шаге.

Если алгоритм произвести $n^2$ раз, чтобы получить вероятность успеха $1 - {1\over e}$.

Сложность стягивания $-$ $O(n)$ (из вершины может выходить $n$ ребер).

Итоговая вероятность $-$ $O(n\cdot n \cdot n^2) = O(n^4)$

# Оптимизация Штейна алгоритма Каргера

Произведем построение дерева случайных применений алгоритма Каргера:

![Дерево случайных применений алгоритма Каргера](./images/karger.png)

Вероятность того, что выполнив стягивание $n(1 - {1 \over \sqrt{2}})$ раз минимальный разрез не будет потерян равна $ \prod\limits_{i=1}^{n(1 - {1 \over \sqrt{2}})}P(x_i) \geq {n-2 \over n}{n-3 \over n-1}\cdot... \approx {1 \over 2} $

Таким образом, вероятность, что при построении дерева применений алгоритма Каргера решение не будет потеряно равна ${1 \over \log(n)}$.

Оценим сложность построения дерева.

$$ T(n = |V|) = n^2 + 2T({n\over \sqrt{2}}) = n^2 + n^2 + 4T({n\over \sqrt{4}}) \approx n^2 \log(n) $$

Получим общую сложность $O(\log(n)\cdot n^2\log(n)) = O((n\log(n))^2)$

Разрезы применяются в задачах кластеризации.
