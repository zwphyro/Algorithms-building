# Метод ближайшего соседа

На каждом шаге совершать переход по инцидентному ребру наименьшей стоимости, инцедентному еще не посещенной вершине.

Данный метод дает неплохую оценку, но для евклидовых графов данный метод может найти цикл, который в $\log{n}$ раз хуже оптимального.  

# Метод ветвей и границ

Пусть найдено какое-то приближенное решение.
1. Если при для текущего частичного решения длина пути больше, чем в уже найденном приближенном решении, то дальнейшие расширения текущего частичного расширения рассматривать не имеет смысла.

2. Если для вершин, не входящих в текущее частичное решение соимостью $S$ найти стоимость минимального остовного дерева $M$, то расширение данного решения не рассматривается, если стоимость приближенного решения меньше или равна $S+M$.

Рассматривать пары ребер и перебирать другие возможные способы объединения инцидентных им вершин, пытаясь найти лучшее локальное решение.

## Имитация отжига

Пусть в самом начале при локальном поиске можно с какой-то вероятностью выбрать более плохое решение.

```
s <- какое-нибудь начальное решение  
повторять:  
    s' <- случайное решение из окружения s  
    D <- cost(s') - cost(s)  
    если D < 0:  
        заменить s на s'  
    иначе:  
        заменить s на s' с вероятностью e^(D/T)
```

# Приближенное решение

1. Построить минимальное остовное дерево T
2. Продублировать каждое ребро дерева T
3. В полученном графе найти эйлеров цикл
4. Если в цикле какая-нибудь вершина встретилась повторно $-$ вычеркнуть ее

Данный алгоритм работает на *метрическом неориентированнои полном* графе (должно выполняться неравенство треугольника).
